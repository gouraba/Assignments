#### Part 1 (Continued): Random List Generation and Analysis

**Objective**: Generate multiple lists of random integers and perform basic statistical operations.

**Instructions**:

2.  For each of these 10 lists:
    *   Calculate the **sum** of its elements.
    *   Calculate the **average** of its elements.
3.  Identify and print the list that has the **highest sum** among all 10 generated lists, along with its sum and average.

```python
import random

# Your code for Part 1 (Continued) goes here

# TODO: Generate 10 lists of random integers

lists = [[random.randint(1, 100) for _ in range(5)] for _ in range(10)]
for i in range(0,10):
    print(f"List {i+1}-",end=" ")
    for x in lists[i]:
        print(x,end=" ")
    print()

print()

# TODO: Calculate sum and average for each list
sm=[]
avg=[]
for i, lst in enumerate(lists, start=1):
    s=sum(lst)
    a=s/len(lst)
    sm.append(s)
    avg.append(a)

for i in range(0,10):
    print(f"For list {i+1}- Sum is {sm[i]}, Average is {avg[i]}")

print()

# TODO: Find and print the list with the highest sum
mx=max(sm)
for i in range(0,10):
    if mx==sum(lists[i]):
        print(f"Max sum is of List {i+1}-",end=" ")
        for x in lists[i]:
            print(x,end=" ")
        print()
```

### Part 3: Bonus Problem - Object-Oriented Book Management

**Objective**: Practice object-oriented programming concepts (`class`, `__init__`) by creating a `Book` class and performing operations on a collection of books.

**Scenario**: You are managing a small library and want to organize book information using Python classes.

**Instructions**:
1.  **Define a `Book` class** with an `__init__` method that takes `title` and `author` as arguments and stores them as instance attributes.
2.  **Create a list of `Book` objects** using the provided `book_data`.
3.  **Implement a method** within the `Book` class (or a separate function) to:
    *   Find the book with the **longest title**.
    *   List all books written by a specific author (e.g., 'George Orwell').
4.  **Print** the results in a clear and readable format.

```python
# Pre-code for Part 3
book_data = [
    {'title': '1984', 'author': 'George Orwell'},
    {'title': 'Brave New World', 'author': 'Aldous Huxley'},
    {'title': 'Animal Farm', 'author': 'George Orwell'},
    {'title': 'The Great Gatsby', 'author': 'F. Scott Fitzgerald'},
    {'title': 'To Kill a Mockingbird', 'author': 'Harper Lee'}
]

# Your code for Part 3 goes here

# TODO: Define the Book class
class Book:
    def __init__(self,title,author):
        self.title=title
        self.author=author


# TODO: Create Book objects and perform operations
books=[]
for book in book_data:
    books.append(Book(book['title'],book['author']))

def longest_title(arr):
    mx=arr[0]
    for x in arr:
        if len(x.title)>len(mx.title):
            mx=x
    print(f"The title of longest book is '{mx.title}' with length {len(mx.title)} and the author is '{mx.author}'")

def author_books(arr,name):
    ans=[]
    for x in arr:
        if x.author==name:
            ans.append(x)
    for x in ans:
        print(f"Book is '{x.title}' by '{x.author}'")

# TODO: Print results
longest_title(books)
print()
author_books(books,'George Orwell')
```
